```python
'''
Pipeline 2 
'''
df_reg = df_1.copy()
df_mix = df_1.copy()

def closest(lst, K): 
    return lst[min(range(len(lst)), key = lambda i: abs(lst[i]-K))] 
      

DEGREE = 1 
MIN_POINT_NUMBER = 4
MIN_DIST = 8000

print('With a linear fit on 4 points or more we could')
for column in df_mix.columns:
    if (column in ['y', 'nans', 'galaxy', 'galactic year']):
        continue
    
    for galaxy in df_mix.galaxy.unique():
        tmp = df_reg.loc[df_mix.galaxy == galaxy, ['galactic year', 'galaxy', column]]
        tmp = tmp.dropna()
        tmp_dist = []
        for x, y in itertools.combinations(tmp['galactic year'], r=2):
            tmp_dist.append(abs(x-y))
            
        if ((len(tmp) >= MIN_POINT_NUMBER) and (max(tmp_dist) > MIN_DIST)) :
            z = np.polyfit(tmp['galactic year'], tmp[column], DEGREE)
            p = np.poly1d(z)
            pred = p(df_reg.loc[df_reg.galaxy == galaxy,'galactic year'])
            std = statistics.stdev(tmp[column])
            pred = pred + np.random.normal(0, std, len(pred))
            max_ = max(tmp[column], key = abs)
            too_far = np.array([abs(max_ - elem) > 5*std for elem in pred])
            pred = [elem if elem in pred[~too_far] else closest(pred[~too_far], elem) for elem in pred]
            df_reg.loc[df_reg.galaxy == galaxy, 'fit {0}'.format(column)] = pred

    c = 'fit {0}'.format(column)
    df_mix.loc[df_mix[column].isna(), column] = df_reg.loc[df_mix[column].isna(),c]
    print('reduce the Nas of {0} to {1}'.format(column,sum(df_mix[column].isna())))
    
    
```
